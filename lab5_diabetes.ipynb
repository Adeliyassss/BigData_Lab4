# KNeighborsRegressor

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.neighbors import KNeighborsRegressor

data = load_diabetes()
X, y = data.data, data.target
# Создание и обучение модели KNeighborsRegressor
# Количество соседей - 5
knn_regressor = KNeighborsRegressor(n_neighbors=5)
knn_regressor.fit(X_train, y_train)

# Предсказание на тестовом наборе
y_pred = knn_regressor.predict(X_test)

# Визуализация результатов
plt.scatter(y_test, y_pred)
plt.xlabel('Фактические значения')
plt.ylabel('Предсказанные значения')
plt.title('Диаграмма рассеяния для KNeighborsRegressor')
plt.show()
```

```python
X = diabetes_dataset.data[:, :2]
y = diabetes_dataset.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

training_accuracy = []
test_accuracy = []

neighbors_settings = range(1, 11)

for n_neighbors in neighbors_settings:
    reg = KNeighborsRegressor(n_neighbors=n_neighbors)
    reg.fit(X_train, y_train)
    training_accuracy.append(reg.score(X_train, y_train))
    test_accuracy.append(reg.score(X_test, y_test))

# Строим график
plt.plot(neighbors_settings, training_accuracy, label="R^2 на обучающем наборе")
plt.plot(neighbors_settings, test_accuracy, label="R^2 на тестовом наборе")
plt.ylabel("R^2")
plt.xlabel("Количество соседей")
plt.legend()
plt.show()
```

```python
import mglearn
import numpy as np

# Анализ KNeighborsRegressor
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
line = np.linspace(-3, 3, 1000).reshape(-1, 10)
line = line[:, :2]  # Используем только первые два признака

for n_neighbors, ax in zip([1, 3, 9], axes):
    reg = KNeighborsRegressor(n_neighbors=n_neighbors)
    reg.fit(X_train[:, :2], y_train)  # Используем только первые два признака из обучающего набора
    ax.plot(line, reg.predict(line[:, :2]))  # Также используем только первые два признака для предсказаний
    ax.plot(X_train[:, 0], y_train, '^', c=mglearn.cm2(0), markersize=8)
    ax.plot(X_test[:, 0], y_test, 'v', c=mglearn.cm2(1), markersize=8)
    
    ax.set_title(
        "{} neighbor(s)\n train score: {:.2f} test score: {:.2f}".format(
            n_neighbors, reg.score(X_train[:, :2], y_train),
            reg.score(X_test[:, :2], y_test)))
    ax.set_xlabel("Признак 1")
    ax.set_ylabel("Целевая переменная")
    axes[0].legend(["Прогнозы модели", "Обучающие данные/ответы", "Тестовые данные/ответы"], loc='best')
```

## Линейная модель регрессии (метод наименьших квадратов)

```python
from sklearn.linear_model import LinearRegression

X, y = load_diabetes(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
lr = LinearRegression().fit(X_train, y_train)
```

```python
print("Правильность на обучающем наборе: {:.2f}".format(lr.score(X_train, y_train)))
print("Правильность на тестовом наборе: {:.2f}".format(lr.score(X_test, y_test)))
```

```python
print("lr.coef_: {}".format(lr.coef_))
print("lr.intercept_: {}".format(lr.intercept_))
```

```python
data = load_diabetes()
X, y = data.data[:, 2], data.target  

reg = LinearRegression()
X = X.reshape(-1, 1) 
reg.fit(X, y)

# Предсказание на данных для построения линии регрессии
y_pred = reg.predict(X)

plt.scatter(X, y, label='Данные', color='blue')
plt.plot(X, y_pred, label='Линейная регрессия', color='red')
plt.xlabel('Признак')
plt.ylabel('Целевая переменная')
plt.legend()
plt.title('Диаграмма линейной регрессии')
plt.show()
```

## Гребневая регрессия

```python
from sklearn.linear_model import Ridge

ridge = Ridge().fit(X_train, y_train)
print("Правильность на обучающем наборе: {:.2f}".format(ridge.score(X_train, y_train)))
print("Правильность на тестовом наборе: {:.2f}".format(ridge.score(X_test, y_test)))
```

```python
ridge10 = Ridge(alpha=10).fit(X_train, y_train)
print("Правильность на обучающем наборе: {:.2f}".format(ridge10.score(X_train, y_train)))
print("Правильность на тестовом наборе: {:.2f}".format(ridge10.score(X_test, y_test)))
```

```python
ridge01 = Ridge(alpha=0.1).fit(X_train, y_train)
print("Правильность на обучающем наборе: {:.2f}".format(ridge01.score(X_train, y_train)))
print("Правильность на тестовом наборе: {:.2f}".format(ridge01.score(X_test, y_test)))
```

```python
plt.plot(ridge.coef_, 's', label="Гребневая регрессия alpha=1")
plt.plot(ridge10.coef_, '^', label="Гребневая регрессия alpha=10")
plt.plot(ridge01.coef_, 'v', label="Гребневая регрессия alpha=0.1")
plt.plot(lr.coef_, 'o', label="Линейная регрессия")
plt.xlabel("Индекс коеффициента")
plt.ylabel("Оценка коеффициента")
plt.hlines(0, 0, len(lr.coef_))
plt.ylim(-400, 400)
plt.legend()
plt.show()
```

```python
mglearn.plots.plot_ridge_n_samples()
```

* Первые два блока кода выводят правильность (R ^2) моделей гребневой регрессии с разными значениями alpha. Высокое значение R ^2 на тестовом наборе данных может указывать на хорошую обобщающую способность модели.
*  График коэффициентов позволяет визуально сравнить влияние разных значений alpha на коэффициенты регрессии. При увеличении alpha коэффициенты сжимаются к нулю, что является проявлением регуляризации. Выбор оптимального значения alpha зависит от конкретной задачи и может потребовать кросс-валидации.
*  Важно выбирать значение alpha, которое достигает баланса между хорошей обобщающей способностью модели и уменьшением риска переобучения.

# Лассо

```python
from sklearn.linear_model import Lasso

lasso = Lasso().fit(X_train, y_train)
print("Правильность на обучающем наборе: {:.2f}".format(lasso.score(X_train, y_train)))
print("Правильность на тестовом наборе: {:.2f}".format(lasso.score(X_test, y_test)))
print("Количество использованных признаков: {}".format(np.sum(lasso.coef_ !=0)))
```

```python
lasso001 = Lasso(alpha=0.01, max_iter=100000).fit(X_train, y_train)
print("Правильность на обучающем наборе: {:.2f}".format(lasso001.score(X_train, y_train)))
print("Правильность на тестовом наборе: {:.2f}".format(lasso001.score(X_test, y_test)))
print("Количество использованных признаков: {}".format(np.sum(lasso001.coef_ !=0)))
```

```python
lasso0001 = Lasso(alpha=0.0001, max_iter=100000).fit(X_train, y_train)
print("Правильность на обучающем наборе: {:.2f}".format(lasso0001.score(X_train, y_train)))
print("Правильность на тестовом наборе: {:.2f}".format(lasso0001.score(X_test, y_test)))
print("Количество использованных признаков: {}".format(np.sum(lasso0001.coef_ !=0)))
```

```python
plt.plot(lasso.coef_, 's', label="Лассо alpha=1")
plt.plot(lasso001.coef_, '^', label="Лассо alpha=0.01")
plt.plot(lasso.coef_, 'v', label="Лассо alpha=0.0001")
plt.plot(ridge01.coef_, 'o', label="Гребневая регрессия alpha=0.1")
plt.legend(ncol=2, loc=(0,1.05))
plt.xlabel("Индекс коеффициента")
plt.ylabel("Оценка коеффициента")
plt.ylim(-400, 400)
plt.hlines(0, 0, len(ridge01.coef_))
plt.show()
```

* Lasso (L1-регуляризация) применяет регуляризацию, которая склоняется к обнулению некоторых коэффициентов регрессии, что может привести к отбору признаков.
* С уменьшением значения alpha увеличивается количество ненулевых коэффициентов, и модель может становиться более гибкой.
* Модель гребневой регрессии (Ridge) не обнуляет коэффициенты и обеспечивает более непрерывные оценки признаков.
* Выбор оптимального значения alpha зависит от конкретной задачи и требований к модели. Настройка alpha позволяет находить баланс между точностью и сложностью модели.
* График позволяет визуально сравнить оценки коэффициентов для разных моделей и значения alpha.
